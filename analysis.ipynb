```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the DataFrame by Year and Day, and calculate the count
yearly_day_count = df.groupby(['Year', 'Day']).size().reset_index(name='Count')

# Plot the bar plot
sns.barplot(x='Day', y='Count', hue='Year', data=yearly_day_count)

# Set labels and title
plt.xlabel('Day')
plt.ylabel('Count')
plt.title('Number of Times Each Day Was Performed (By Year)')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the DataFrame by Day and Exercise, and calculate the count
exercise_count = df.groupby(['Day', 'Exercise']).size().reset_index(name='Count')

# Sort the exercise_count DataFrame by Count in descending order within each Day
exercise_count_sorted = exercise_count.groupby('Day').apply(lambda x: x.sort_values('Count', ascending=False)).reset_index(drop=True)

# Get the top 10 exercises for each day
top_10_exercises = exercise_count_sorted.groupby('Day').head(10)

# Plot the bar plots for each day
g = sns.FacetGrid(top_10_exercises, col='Day', col_wrap=3, height=4, sharey=False)
g.map(sns.barplot, 'Count', 'Exercise', palette='viridis', ci=None)
g.set_titles('{col_name}')
g.set_axis_labels('Count', 'Exercise')

# Adjust the spacing between subplots
plt.subplots_adjust(top=0.9, hspace=0.5)

# Set the overall title
plt.suptitle('Top 10 Exercises Performed for Each Day')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count occurrences of specific words in 'Exercise' column
bb_count = df['Exercise'].str.contains('BB', case=False).sum()
db_count = df['Exercise'].str.contains('DB', case=False).sum()
cable_count = df['Exercise'].str.contains('Cable', case=False).sum()

# Create a DataFrame for the counts
word_counts = pd.DataFrame({'Word': ['BB', 'DB', 'Cable'],
                            'Count': [bb_count, db_count, cable_count]})

# Plot the bar plot
sns.barplot(x='Word', y='Count', data=word_counts)

# Set labels and title
plt.xlabel('Word')
plt.ylabel('Count')
plt.title('Type of weights')

# Show the plot
plt.show()
```
